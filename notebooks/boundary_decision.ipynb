```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the live inferences
df = pd.read_csv("live_shadow_run.csv")

# Ensure fake_prob is a float
df['fake_prob'] = df['new_fake_prob'].astype(float)

print("\n--- SHADOW RUN STATS ---")
print(f"Total Live Reviews Analyzed: {len(df)}")
print(f"Mean Fake Probability: {df['fake_prob'].mean():.4f}")
print(f"Median Fake Probability: {df['fake_prob'].median():.4f}")

# 2. Percentile Mapping (Where does the data actually sit?)
print("\n--- CONFIDENCE PERCENTILES ---")
percentiles = [50, 75, 85, 90, 95, 98, 99]
for p in percentiles:
    val = np.percentile(df['fake_prob'], p)
    print(f"{p}th Percentile: {val:.4f} Fake Score")

# 3. Plot the Probability Density
plt.figure(figsize=(10, 6))
sns.histplot(df['fake_prob'], bins=50, kde=True, color='darkred')
plt.title("Probability Density of 'Fake' Scores on Live Amazon Data", fontweight='bold')
plt.xlabel("DeBERTa Fake Probability (0.0 to 1.0)")
plt.ylabel("Number of Reviews")
plt.axvline(0.5, color='black', linestyle='--', label='Standard 0.5 Threshold')

# 4. Find the Natural "Valleys"
# If there is a bimodal distribution (a big hump of humans, a gap, and a small hump of bots),
# this plot will show you exactly where the threshold should go.
plt.legend()
plt.tight_layout()
plt.savefig("live_distribution_analysis.png", dpi=150)
print("\nSaved density plot to 'live_distribution_analysis.png'")

# 5. Extract the Extremes for Manual Ground-Truthing
print("\n--- TOP 5 HIGHEST FAKE SCORES (The undeniable bots?) ---")
for text in df.sort_values('fake_prob', ascending=False)['review_text'].head(5):
    print(f"- {text[:150]}...")

print("\n--- 5 REVIEWS NEAR THE 75th PERCENTILE (The terse humans?) ---")
p75_val = np.percentile(df['fake_prob'], 75)
# Grab reviews that are right around the 75th percentile mark
middle_zone = df[(df['fake_prob'] > p75_val - 0.02) & (df['fake_prob'] < p75_val + 0.02)]
for text in middle_zone['review_text'].head(5):
    print(f"- {text[:150]}...")
```
